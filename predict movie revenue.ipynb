# Notebook Imports and Packages

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
data.describe()
```

```python
x = pd.DataFrame(data, columns=['production_budget_usd'])
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
# type(x)
```

```python
plt.figure(figsize=(12,8))

plt.scatter(x,y, alpha=0.4)

plt.title('Film Cost Vs Global Revenue', fontsize =20)
plt.xlabel('Production Budget', fontsize =16)
plt.ylabel('Worldwide Gross', fontsize =16)

plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```

## Linear Regression:
<p>Used to predict the value of a variable best on the value of another variable </p>

```python
regression = LinearRegression()
regression.fit(x, y)
```

### <li>slope coefficient:</li>

```python
regression.coef_
```

### <li>Intercept:</li>

```python
regression.intercept_
```

```python
plt.figure(figsize=(12,8))

plt.scatter(x,y, alpha=0.4)
plt.plot(x, regression.predict(x), color='red', linewidth=4)

plt.title('Film Cost Vs Global Revenue', fontsize =20)
plt.xlabel('Production Budget', fontsize =16)
plt.ylabel('Worldwide Gross', fontsize =16)

plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```

## $R^2$ (Goodness of Fit):
<p>measure of how close the data are to the fitted regression line</p>

```python
regression.score(x, y)
```
